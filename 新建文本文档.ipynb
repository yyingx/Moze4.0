## 调试 sniff_and_load_data 函数

下面是从主脚本中提取的 `sniff_and_load_data` 函数，用于调试和测试。

```python
# 导入必要的库
import pandas as pd
from pathlib import Path

# 列名映射字典（从主脚本复制）
COLUMN_MAPPING = {
    '交易时间': '交易时间', '交易创建时间': '交易时间', '付款时间': '交易时间',
    '交易类型': '交易类型', '交易分类': '交易类型', '交易对方': '交易对方',
    '商品': '商品', '商品名称': '商品', '收/支': '收/支',
    '金额(元)': '金额(元)', '金额（元）': '金额(元)', '金额': '金额(元)',
    '支付方式': '支付方式', '收/付款方式': '支付方式',
    '当前状态': '当前状态', '交易状态': '当前状态',
    '备注': '备注', '交易单号': '交易单号', '交易订单号': '交易单号',
    '商户单号': '商户单号', '商家订单号': '商户单号', '商品说明': '商品'
}

print("✅ 导入完成，COLUMN_MAPPING 已加载")
```

```python
def sniff_and_load_data(file_path: Path):
    """
    读取支付宝/微信账单文件
    
    参数:
        file_path: 账单文件路径 (Path 对象)
    
    返回:
        DataFrame 或 None
    """
    print(f"📁 读取文件: {file_path.name}")
    print(f"   后缀: {file_path.suffix.lower()}")
    
    ftype, df = None, None
    
    try:
        # === 步骤1: 根据文件类型读取 ===
        if file_path.suffix.lower() == '.csv':
            print("   类型: CSV → 支付宝账单")
            print("   参数: header=24, encoding=gb18030")
            df = pd.read_csv(file_path, header=24, encoding='gb18030')
            ftype = "AliPay"
        elif file_path.suffix.lower() == '.xlsx':
            print("   类型: XLSX → 微信账单")
            print("   参数: header=16, engine=openpyxl")
            df = pd.read_excel(file_path, header=16, engine='openpyxl')
            ftype = "WeChat"
        else:
            print(f"   ❌ 不支持的文件类型: {file_path.suffix}")
            return None
            
        if df is None:
            print("   ❌ DataFrame 为空")
            return None
        
        print(f"   ✅ 读取成功: {len(df)} 行 × {len(df.columns)} 列")
        
        # === 步骤2: 清理列名 ===
        print(f"\n🔧 原始列名: {list(df.columns)[:5]}...")  # 只显示前5个
        
        df.columns = df.columns.astype(str).str.strip().str.replace(
            '[（(]', '(', regex=True).str.replace('[）)]', ')', regex=True)
        
        print(f"   清理后: {list(df.columns)[:5]}...")
        
        # === 步骤3: 列名映射 ===
        df.rename(columns=COLUMN_MAPPING, inplace=True)
        print(f"   映射后: {list(df.columns)[:5]}...")
        
        # === 步骤4: 检查必需列 ===
        if '金额(元)' not in df.columns:
            print(f"   ❌ 缺少 '金额(元)' 列")
            print(f"   当前列: {list(df.columns)}")
            return None
        
        # === 步骤5: 处理金额列 ===
        print(f"\n💰 金额列处理:")
        print(f"   原始类型: {df['金额(元)'].dtype}")
        print(f"   示例值: {df['金额(元)'].head(3).tolist()}")
        
        df['金额(元)'] = pd.to_numeric(
            df['金额(元)'].astype(str).str.replace(r'[¥,]', '', regex=True).str.strip(),
            errors='coerce'
        ).fillna(0)
        
        df['金额'] = df['金额(元)']
        
        print(f"   处理后类型: {df['金额(元)'].dtype}")
        print(f"   处理后示例: {df['金额(元)'].head(3).tolist()}")
        
        # === 步骤6: 添加标签 ===
        df['_source_tag'] = '#WechatPay' if ftype == "WeChat" else '#AliPay'
        
        print(f"\n✅ 处理完成 (类型: {ftype})")
        print(f"   总行数: {len(df)}")
        print(f"   总金额: {df['金额'].sum():.2f} 元")
        
        return df
        
    except Exception as e:
        print(f"\n❌ 读取失败: {e}")
        import traceback
        traceback.print_exc()
        return None

print("✅ 函数 sniff_and_load_data 已定义")
```
